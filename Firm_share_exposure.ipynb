```python
import pandas as pd

# Assuming df is your original DataFrame
df = federal_final.copy()

# Columns to be processed
funding_sources = ['dod', 'doe', 'nasa', 'nsf', 'hhs']
cols = ['A2', 'A0', 'A4', 'A6', 'B0', 'B2', 'B3', 'B4', 'B6', 'B8', 'C0', 'C1', 'C2', 'C3', 'C4', 'C7',
        'D0', 'D1', 'D2', 'E0', 'E2', 'F0', 'F1', 'F2', 'F4', 'G0', 'G1', 'G2', 'G6', 'H0', 'Y0']

# Create an empty DataFrame to hold all results
final_federal_df = pd.DataFrame()

# Loop through each funding source, calculate product, and sum by year
for source in funding_sources:
    # Compute product of source with each column in cols
    df_product = df[cols].mul(df[source], axis=0)
    df_product['year'] = df['year']
    df_product['source'] = source  # Add a column indicating the funding source

    # Group by 'year' and sum all columns
    df_collapsed = df_product.groupby(['year', 'source']).sum().reset_index()

    # Append results to the final DataFrame
    final_federal_df = pd.concat([final_federal_df, df_collapsed], ignore_index=True)

# Sorting by year and source for better readability
final_federal_df.sort_values(by=['year', 'source'], inplace=True)
```

```python
final_federal_df.tail(20)
```

```python
import pandas as pd
final_federal_df.set_index(['source', 'year'], inplace=True)
# Assuming 'final_federal_df' is already loaded and has 'source' and 'year' as a multi-index
# Define the columns we're interested in
cols = ['A2', 'A0', 'A4', 'A6', 'B0', 'B2', 'B3', 'B4', 'B6',
        'B8', 'C0', 'C1', 'C2', 'C3', 'C4', 'C7', 'D0', 'D1', 'D2', 'E0', 'E2',
        'F0', 'F1', 'F2', 'F4', 'G0', 'G1', 'G2', 'G6', 'H0', 'Y0']

# Function to sum data from t-5 to t-10
def sum_data_from_past(row, df, start_years_ago=10, end_years_ago=5):
    source, current_year = row['source'], row['year']
    years = range(current_year - start_years_ago, current_year - end_years_ago + 1)
    data_sum = pd.Series([0] * len(cols), index=cols)

    for year in years:
        try:
            data_sum += df.loc[(source, year), cols].fillna(0)  # Adding fillna(0) to handle missing data
        except KeyError:
            continue  # Skip years where data does not exist

    return data_sum

# Apply the function to each row in a new DataFrame that includes 'year' and 'source'
new_df_federal = final_federal_df.reset_index()
new_df_federal[cols] = new_df_federal.apply(sum_data_from_past, axis=1, df=final_federal_df)

# Reset index if needed or perform additional cleanup
#new_df.set_index(['source', 'year'], inplace=True)

# Show the result
print(new_df_federal.head(10))
```

```python
import pandas as pd
import numpy as np
final_federal_df = new_df_federal.copy()
final_df = copy_of_new_df.copy()



# Columns for vector calculations
cols = ['A2', 'A0', 'A4', 'A6', 'B0', 'B2', 'B3', 'B4', 'B6', 'B8', 'C0', 'C1', 'C2', 'C3', 'C4', 'C7',
        'D0', 'D1', 'D2', 'E0', 'E2', 'F0', 'F1', 'F2', 'F4', 'G0', 'G1', 'G2', 'G6', 'H0', 'Y0']

# Preparing the new DataFrame
results = []

# Loop through each unique year, source, and gvkeyuo
for year in set(final_df['year']).intersection(set(final_federal_df['year'])):
    df_year = final_df[final_df['year'] == year]
    fed_year = final_federal_df[final_federal_df['year'] == year]

    for index1, row1 in df_year.iterrows():
        gvkeyuo = row1['gvkeyuo']
        p_gvkeyuo = row1[cols].values.astype(float)

        for index2, row2 in fed_year.iterrows():
            source = row2['source']
            p_source = row2[cols].values.astype(float)

            # Calculate dot product and magnitudes
            dot_product = np.dot(p_gvkeyuo, p_source)
            magnitude_gvkeyuo = np.sqrt(np.sum(p_gvkeyuo**2))
            magnitude_source = np.sqrt(np.sum(p_source**2))

            # Calculate cosine similarity
            cosine_similarity = dot_product / (magnitude_gvkeyuo * magnitude_source)

            # Append results
            results.append({'year': year, 'source': source, 'gvkeyuo': gvkeyuo, 'c': cosine_similarity})

# Create DataFrame from results
results_df = pd.DataFrame(results, columns=['year', 'source', 'gvkeyuo', 'c'])

# Displaying part of the DataFrame
results_df.head()
```

```python
import pandas as pd
import numpy as np

# Assuming you've already generated results_df with the 'c' values calculated

# Step 1: Replace NaN values in 'c' with 0
results_df['c'] = results_df['c'].fillna(0)

# Step 2: Calculate the total sum of 'c' for each 'gvkeyuo' and 'year'
sum_c = results_df.groupby(['gvkeyuo', 'year'])['c'].transform('sum')

# Step 3: Normalize 'c' by dividing each 'c' by the total sum for its 'gvkeyuo' and 'year'
results_df['c_not_normalized'] = results_df['c']
results_df['c'] = results_df['c'] / sum_c

# Displaying part of the normalized DataFrame to check the results
results_df['c'] = results_df['c'].fillna(0)
results_df['c_not_normalized'] = results_df['c_not_normalized'].fillna(0)
```
